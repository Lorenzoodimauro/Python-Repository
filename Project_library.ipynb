```python
# Creo una lista vuota iniziale di elementi in questo caso di libri.
lista_libri = []

# Creo una funzione che aggiunge all'interno della lista dei libri.

def registra_libro(titolo, autore, totale_pagine):
    libro = {   "titolo": titolo,
                "autore": autore,
                "totale_pagine": totale_pagine
            }
    
    lista_libri.append(libro)
       
registra_libro("Harry Potter e il Calice di Fuoco", "J.K. Rowling", 636)
registra_libro("Harry Potter e il Prigioniero di Azkaban", "J.K. Rowling", 317)
registra_libro("Harry Potter e la Pietra Filosofale", "J.K. Rowling", 223)
registra_libro("Il Signore degli Anelli - La Compagnia dell'Anello", "J.R.R. Tolkien", 423)
registra_libro("Il Signore degli Anelli - Le Due Torri", "J.R.R. Tolkien", 352)
registra_libro("Il Signore degli Anelli - Il Ritorno del Re", "J.R.R. Tolkien", 416)
registra_libro("Harry Potter e i Doni della Morte", "J.K. Rowling", 607)


# Creo una funzione che permette di visualizzare gli elementi aggiunti alla lista.

def visualizza_libri():
    print("\nLISTA LIBRI ATTUALE")
    
    for libro in lista_libri:
        print(f"- {libro['titolo']}-, -autore: {libro['autore']}-, -totale pagine: {libro['totale_pagine']}-")

# Testo la funzione.

visualizza_libri()

# Creo una funzione di ricerca flessibile che da come risultato il libro che l'utente sta cercando.

# Utilizzo il valore speciale None per rendere piu flessibile la ricerca del libro, l'utente puo' cercare il suo libro anche solo digitando il titolo.

def cerca_libro(titolo, autore=None, totale_pagine=None):
    for libro in lista_libri:
        
        if (libro["titolo"] == titolo and 
            (autore is None or libro["autore"] == autore) and 
            (totale_pagine is None or libro["totale_pagine"] == totale_pagine)):

            # Controllo tempo di lettura basato sulle pagine
            if 300 <= libro["totale_pagine"] <= 400:
                print("Tempo di lettura: 2 settimane.")
           
            elif libro["totale_pagine"] >= 600:
                print("Libro molto lungo, tempo lettura: 1 mese.")
            
            elif libro["totale_pagine"] <= 250:
                print("Libro corto, tempo lettura: 1 settimana.")

            return libro  

    print("Libro non disponibile.")
    return None


# Testo la funzione.

print(cerca_libro("Harry Potter e il Prigioniero di Azkaban"))
```

```python
# Creo una funzione che restituisce il numero di libri presenti nella lista e il numero di libri scritti da ogni autore.
def genera_statistiche():
    contatore_libri = len(lista_libri)  # Metodo più diretto per contare i libri
    print(f"Il totale dei libri presenti nella lista: {contatore_libri}")

    statistiche_autore = {}

    # Controllo per evitare errori se la lista è vuota
    if len(lista_libri) > 0:
        libro_max = lista_libri[0]  # Imposto il primo libro come riferimento
        libro_min = lista_libri[0]

        # Ciclo che scorre tutti i libri e calcola statistiche
        for libro in lista_libri:
            autore = libro["autore"]
            
            # Conteggio libri per autore
            if autore in statistiche_autore:
                statistiche_autore[autore] += 1
            else:
                statistiche_autore[autore] = 1

            # Trova il libro con più pagine
            if libro["totale_pagine"] > libro_max["totale_pagine"]:
                libro_max = libro  # Manteniamo tutto il libro, non solo il numero di pagine

            # Trova il libro con meno pagine
            if libro["totale_pagine"] < libro_min["totale_pagine"]:
                libro_min = libro  # Manteniamo tutto il libro, non solo il numero di pagine

        # Stampa dei risultati
        print("\nNumero di libri per autore:")
        for autore, numero_libri in statistiche_autore.items():
            print(f"- {autore} ha scritto: {numero_libri} libri")

        print("\nIl libro con più pagine:")
        print(f"- Titolo: {libro_max['titolo']}, Autore: {libro_max['autore']}, Pagine: {libro_max['totale_pagine']}")

        print("\nIl libro con meno pagine:")
        print(f"- Titolo: {libro_min['titolo']}, Autore: {libro_min['autore']}, Pagine: {libro_min['totale_pagine']}")
    
    else:
        print("La lista dei libri è vuota, impossibile calcolare le statistiche.")


# Eseguo la funzione

genera_statistiche()


# Creo una Funzione che filtra in modo avanzato la ricerca del libro da parte dell'utente>

def filtro_avanzato(autore, min_pagine, max_pagine):
    
    libri_trovati = []
    
    for libro in lista_libri:
         if min_pagine <= libro['totale_pagine'] <= max_pagine and libro['autore'] == autore:
            libri_trovati.append(libro) 
            
    if not libri_trovati:
        print("Nessun libro trovato.")
    else:
        print("Libri trovati:")
        for libro in libri_trovati:
                print(f"- Titolo: {libro['titolo']}, Autore: {libro['autore']}, Pagine: {libro['totale_pagine']}")
        
    


# Testo la funzione

filtro_avanzato("J.R.R. Tolkien", 200,600 )
```

```python
while True:
    print("\nMENU PRINCIPALE")
    print("1. Aggiungere un libro")
    print("2. Visualizzare tutti i libri")
    print("3. Cercare un libro")
    print("4. Filtrare libri per autore e pagine")
    print("5. Visualizzare statistiche")
    print("0. Esci")

    scelta = input("Scegli un'opzione: ")

    if scelta == "1":
        # Aggiungere un libro
        titolo = input("Inserisci il titolo del libro: ")
        autore = input("Inserisci l'autore del libro: ")
        pagine = int(input("Inserisci il numero di pagine: "))
        registra_libro(titolo, autore, pagine)  

    elif scelta == "2":
        # Mostra tutti i libri
        visualizza_libri()

    elif scelta == "3":
        # Cerca un libro con `None` per i campi opzionali
        print("\nCERCA UN LIBRO")
        titolo = input("Inserisci il titolo del libro da cercare (o lascia vuoto): ")
        autore = input("Inserisci l'autore (o lascia vuoto): ")
        pagine = input("Inserisci il numero di pagine (o lascia vuoto): ")

        # Se il campo è vuoto, assegniamo None per indicare che non vogliamo filtrare su quel parametro
        titolo = titolo if titolo.strip() else None
        autore = autore if autore.strip() else None
        pagine = int(pagine) if pagine.strip() else None  

        risultato = cerca_libro(titolo, autore, pagine)

        if risultato:
            print("\nLibro trovato:")
            print(f"- Titolo: {risultato['titolo']}")
            print(f"- Autore: {risultato['autore']}")
            print(f"- Pagine: {risultato['totale_pagine']}")
        else:
            print("Libro non trovato.")

    elif scelta == "4":
        # Filtra libri in base a autore e pagine
        print("\nFILTRA LIBRI")
        autore = input("Inserisci l'autore da filtrare: ")
        min_pagine = int(input("Inserisci il numero minimo di pagine: "))  
        max_pagine = int(input("Inserisci il numero massimo di pagine: "))
        filtro_avanzato(autore, min_pagine, max_pagine)

    elif scelta == "5":
        # Genera statistiche sui libri
        genera_statistiche()

    elif scelta == "0":
        print("Uscita dal programma. Arrivederci!")
        break  # Esce dal menu e termina il programma

    else:
        print("Scelta non valida. Riprova.")

    input("\nPremi INVIO per tornare al menu...")
```
